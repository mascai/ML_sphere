```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

```python
df1 = pd.read_csv('./dataset1.csv', sep='\t')
df2 = pd.read_csv('./dataset2.csv', sep='\t')
df3 = pd.read_csv('./dataset3.csv', sep='\t')
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
data = []
for str in open('dataset1.csv'):
    str_x, str_y = str.strip().split('\t')
    data.append([float(str_x), float(str_y)])
```

```python
clast_num = 5
kmeans = KMeans(n_clusters=clast_num, random_state=0).fit(data)
ans = kmeans.predict(data)

agglo = AgglomerativeClustering(n_clusters=3, affinity='euclidean').fit(data)
ans1 = agglo.fit_predict(data)

dbs = DBSCAN().fit(data)
ans2 = DBSCAN().fit_predict(data)
```
